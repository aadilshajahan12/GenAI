```python
from langchain_community.document_loaders import PDFPlumberLoader
loader=PDFPlumberLoader("Final TOC.pdf")
docs=loader.load()
```

```python
print('no . of pages in pdf',len(docs))
```

```python
docs
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=70)
chunks=splitter.split_documents(docs)
```

```python
for i in chunks:
    print(i.page_content)
    print('*'*100)
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
embedding=HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
```

```python
from langchain_community.vectorstores import FAISS
faiss=FAISS.from_documents(chunks,embedding)
```

```python
retriever=faiss.as_retriever(search_kwargs={'k':3})
```

```python
retrieved_docs=retriever.invoke('what is GAN')
```

```python
retrieved_docs
```

```python
import ollama
LANGUAGE_MODEL = 'hf.co/bartowski/Llama-3.2-1B-Instruct-GGUF'
```

```python
query=input('user:')
retrieve_docs=retriever.invoke(query)
retrieve_docs
```

```python
for i in retrieve_docs:
    print(i.page_content)
    print("*"*100)
```

```python
instructor_prompt=f"""you are a helpful chatbot.
use only the given context of text to answer.
do not create new information for yourself. {''.join(know.page_content for know in retrieve_docs)}"""
```

```python
stream=ollama.chat(
    model=LANGUAGE_MODEL,
    messages=[{'role':'system','content':instructor_prompt},
              {'role':'user','content':query}],stream=True
)
```

```python
for i in stream:
    print(i['message']['content'],end='')
```
