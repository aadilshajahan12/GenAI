```python
from crewai import Agent,Task,Crew,LLM
```

```python
local_llm=LLM(model='ollama/llama3.2:1b')
```

```python
checker=Agent(role='Checker',goal='Identify the urgency of the {topic}',
                 backstory='You are an expert analyser who evaluates the urgency of the message.',llm=local_llm)
```

```python
writer=Agent(role='writer',goal='create mail based on the urgency',
                 backstory='You are a writer who is expertised in preparing response mails',llm=local_llm)
```

```python
check=Task(agent=checker,description='read the request and analyze the amount of urgency',expected_output='label the request as either urgent or not')
```

```python
write=Task(agent=writer,description="from the checker's evaluation, create a mail as the response including the level of urgency",expected_output='response in mail format' )
```

```python
crew=Crew(agents=[checker,writer],tasks=[check,write])
```

```python
result=crew.kickoff(inputs={'topic':'can i get the reports before the meeting?'})
print('*'*100)
print(result)
```
