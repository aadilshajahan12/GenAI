```python
from crewai import Agent,Task,Crew,LLM
```

```python
local_llm=LLM(model='ollama/llama3.2:1b')
```

```python
#Researcher
researcher=Agent(role='Researcher',goal='Identify the major trends in {topic}',
                 backstory='You are a researcher who finds the relevant and upto date informations',llm=local_llm,
                 verbose=True)
```

```python
#Writer
writer=Agent(role='Writer',goal='Summarize the research content and create a report',
                 backstory='You are a professional technical writer expertised in generating reports',llm=local_llm,
                 verbose=True)
```

```python
research=Task(description='search and analyze latest trends about latest topic',
              expected_output='bulleted list of top 3 trends found',
              agent=researcher)
```

```python
write=Task(description="create a 2 paragraph summary based on researcher's findings",
              expected_output='a clean formatted summary report',
              agent=writer)
```

```python
crew=Crew(agents=[researcher,writer],tasks=[research,write])
```

```python
result=crew.kickoff(inputs={'topic':'fifa world cup'})
print('*'*100)
print(result)
```
